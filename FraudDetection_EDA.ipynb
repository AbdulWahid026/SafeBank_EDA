### Selecting Data with loc and iloc Operations

```python
df.iloc[:15,[0,1,2]]
```

```python
df.loc[df['Location']=='New York']
```

```python
df.loc[(df['Transaction_Amount']>5000) & (df['Risk_Score']>0.8)]
```

#### 1. Selecting the last 5 columns for all rows?

##### You can select the last 5 columns using: **`df.iloc[:, -5:]`**.
##### This slices all rows and only the last five columns without naming them.

#### 2. Difference between `.loc` and `.iloc`?

##### **`.loc`** selects data by **labels** (row/column names).
##### **`.iloc`** selects data by **integer positions** (row/column numbers).

*Example:*
`df.loc[0:3, 'Age']` uses labels, while `df.iloc[0:3, 2]` uses index positions.

#### 3. Why indexing is important in large datasets?

Indexing makes data access, filtering, and lookup **much faster**.
It improves efficiency and reduces computation time when working with large datasets.


### GroupBy Operations

```python
df.groupby('Location')['Transaction_Amount'].mean()
```

```python
df[df['Fraud_Label'] == 1].groupby('Transaction_Type')['Fraud_Label'].count()
```

```python
df.groupby(['Device_Type', 'Card_Type'])['Risk_Score'].mean()
```

```python
df.groupby(['Device_Type', 'Card_Type'])['Risk_Score'] \
  .mean() \
  .sort_values(ascending=False)
```
