#### 2. Difference between `.loc` and `.iloc`?

##### **`.loc`** selects data by **labels** (row/column names).
##### **`.iloc`** selects data by **integer positions** (row/column numbers).

*Example:*
`df.loc[0:3, 'Age']` uses labels, while `df.iloc[0:3, 2]` uses index positions.

##### **1. Why is it not always a good idea to remove all outliers? (Example: Transaction_Amount)**

Because some outliers represent **real, important events**, such as unusually large but valid transactions.
Removing them may hide fraud patterns or distort financial behavior.

##### **2. Handling extremely high but legitimate corporate Transaction_Amount**

Use **capping or log transformation**, not removal, because the value is real and must be preserved.
Transformation reduces skew, while capping prevents extreme influence on models.

##### **3. Why simple IQR fails for correlated outliers & how to detect compound outliers?**

IQR looks at each column separately, so it misses cases where outliers occur **only when variables combine**.
Use **multivariate methods** like Mahalanobis distance, clustering, or Isolation Forest to detect such compound outliers.


### Missing Values Detection & Solutions

```python
missing = df.isnull().sum()
percent = (df.isnull().mean() * 100)
print("Num of Missing Values:\n", missing)
print("\nPercentage of Missing Values:\n", percent)
```

```python
df['Transaction_Amount']=df['Transaction_Amount'].fillna(df['Transaction_Amount'].mean())
df['Account_Balance']=df['Account_Balance'].fillna(df['Account_Balance'].mean())
```

```python
for col in ['Device_Type', 'Location', 'Merchant_Category']:
    df[col]=df[col].fillna(df[col].mode()[0])
```

#### **4. Strategy for handling missing Timestamp values**

You can **forward-fill (ffill), backward-fill (bfill), or drop** if timestamps are essential.
For time-series analysis, **interpolation or replacing with nearest valid time** is recommended.


##### 1. What happens if you fill missing values with mean in a skewed column?

It can **distort the distribution** because the mean is pulled by extreme values.
Using **median** is better for skewed data.

##### 2. How can you flag missing values for further analysis without filling them?

Create a **binary indicator column** (e.g., `df['Amount_missing'] = df['Amount'].isnull()`).
This preserves missingness information for modeling.

##### 3. Why is handling missing values critical before modeling?

Models cannot handle NaNs, so missing data can cause **errors or poor performance**.
Proper treatment ensures **accurate, unbiased, and stable predictions**.

### Data Binning with cut and qcut

```python
df['TA_cut'] = pd.cut(df['Transaction_Amount'], bins=5)
df['TA_cut']
```

```python
df['TA_qcut'] = pd.qcut(df['Transaction_Amount'], q=4)
df['TA_qcut']
```

```python
df[df['Risk_Score'] > 0.8].groupby('TA_qcut').size().plot(kind='bar')
```

##### 1. Difference between `cut` and `qcut` + when to use each.

`cut` creates **equal-width bins**, while `qcut` creates **equal-frequency (quantile) bins**.
Use `cut` for fixed numeric ranges; use `qcut` when data is skewed and you need balanced groups.

##### 2. How binning helps detect patterns or risky transactions?

Binning groups continuous values into **ranges**, making trends easier to spot.
It reveals which bins contain **more frauds or high-risk behavior**.

##### 3. How cut vs qcut behave with extreme outliers + which is better?

`cut` becomes **distorted** because outliers stretch the bin ranges.
`qcut` is **more stable** since each bin has equal data—better for visualization with outliers.
